```python
import math
import random

def equation_1():
    return (math.log10(5000) + math.exp(2)) / (12 ** 2)

def equation_2():
    return math.sin(math.radians(30)) + math.cos(math.radians(60)) + math.tan(math.radians(45))

def equation_3():
    return math.factorial(6) / math.sqrt(144)

def equation_4():
    return (math.pi ** 2 + math.e ** 3) / math.sqrt(25)

def equation_5():
    return abs(-120) + pow(5, 3) - math.log(100)

def equation_6():
    return ((math.sin(math.radians(45)) * (12 ** 2)) + math.sqrt(64)) / (math.pi * 2.718) + math.log(50)

def equation_7():
    return math.ceil(12.37) + math.floor(99.99)

def equation_8():
    return round(random.random() * 100)

def equation_9():
    return math.sinh(2) + math.cosh(3)

def equation_10():
    return (math.sqrt(81) * math.log10(1000)) / (math.cos(math.radians(60)) + 2 ** 5)

def display_menu():
    print("\n===== MATH EQUATIONS CALCULATOR =====")
    print("1. (log10 + e^x) / (12²)")
    print("2. sin(30°) + cos(60°) + tan(45°)")
    print("3. Factorial + Square Root")
    print("4. π² + e³ / √25")
    print("5. abs() + power - log")
    print("6. Complex expression")
    print("7. Ceil + Floor")
    print("8. Random Number (0–100)")
    print("9. Hyperbolic Functions")
    print("10. Nested Expression")
    print("0. Exit")
    print("=====================================")

def main():
    while True:
        display_menu()

        try:
            choice = int(input("\nEnter your choice (0-10): "))
        except ValueError:
            print("\n❌ Invalid input. Enter numbers only.")
            continue

        if choice == 0:
            print("\n✅ Program Ended.")
            break

        operations = {
            1: equation_1,
            2: equation_2,
            3: equation_3,
            4: equation_4,
            5: equation_5,
            6: equation_6,
            7: equation_7,
            8: equation_8,
            9: equation_9,
            10: equation_10
        }

        if choice in operations:
            result = operations[choice]()
            print(f"\n✅ Result of Equation {choice}: {result}\n")
        else:
            print("\n❌ Invalid option. Try again.")

if __name__ == "__main__":
    main()
```
